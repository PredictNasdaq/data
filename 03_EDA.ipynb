```python
import pandas as pd

# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
data = pd.read_csv('data.csv')

# Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏
data.head()
```

```python
# Display basic statistics
print(data.describe(include='all'))

# Check for missing values
print(data.isnull().sum())

# Visualize the distribution of 'post_time'
import matplotlib.pyplot as plt

data['post_time'] = pd.to_datetime(data['post_time'])
data['post_time'].dt.date.value_counts().sort_index().plot(kind='line', figsize=(12, 6))
plt.title('Number of Posts Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Posts')
plt.show()

# Visualize the distribution of 'author_name'
data['author_name'].value_counts().head(10).plot(kind='bar', figsize=(12, 6))
plt.title('Top 10 Authors by Number of Posts')
plt.xlabel('Author Name')
plt.ylabel('Number of Posts')
plt.show()
```
